```
from copy import deepcopy
################## Non-Simple Graph ####################

### Even the simple graph will have multiple edges and loops after some iterations hence, consider the non-simple graph from the beginning
### Algorithm is based on Proof of Cayley's Formula, leverage the single existence of least Node (here "0")
### G[i][j] = list of edges between i and j
### 0 is the Super Node 
 
### Count of unique spanning trees
C = 0

### Store All the spanning trees 
ans = {}

### Number of nodes in the graph G
n = 4

def add_edge(G,u,v,e):
  if u not in G:
    G[u] = {}
  if v not in G:
    G[v] = {}
  if v not in G[u]:
    G[u][v] = set()
    G[v][u] = set()
  G[u][v].add(e)
  G[v][u].add(e)

def remove_edge(G,u,v,e=None):
  if e:
    G[u][v].remove(e)
    G[v][u].remove(e)
  if len(G[u][v]) == 0 or e == None:
    del G[u][v]
    if u == v : return
    del G[v][u]
    

def collapse(G,v,e):
  Gd = deepcopy(G)
  Gd[0][v].remove(e)
  Gd[v][0].remove(e)
  for node in list(Gd[v]):
    for e in Gd[v][node]:
      add_edge(Gd,0,node,e)
    remove_edge(Gd,v,node)
  del Gd[v]
  return Gd

def print_graph(G):
  for u in G:
    for node in G[u]:
      for e in list(G[u][node]):
        #if u<=node:
        print(u,node)
  print('\n\n')
 
def rec(G,s=[]):

  #print_graph(G)
  global ans,C,n
  
  ## Termination Condition of Backtracking
  if len(s)==n-1:
    print(C,s)
    ans[C]=s
    C+=1
    return

  ## All neighbours
  for node in G[0]:

    ## Loop Case
    if node == 0: continue 

    ## Multiple edges case
    for e in list(G[0][node]):

      ## Collapse node into super node 0
    
        Gd = collapse(G,node,e)

        ## Add e to tree and pass it to recursion
        s.append(e)

        ## Recursively find all trees containing "e"
        rec(Gd,s)

        ## Remove e from tree since everything is not about "e"
        s.remove(e)  
        
        ## We already have covered all the possible trees that contains "e"
        G[0][node].remove(e)
        G[node][0].remove(e)

      

 
G = {}
c = 0
for i in range(n):
  for j in range(i+1,n):
    if i==0 and j==2:continue 
    add_edge(G,i,j,'e'+str(c))
    c+=1
# G[0][2].remove('e1')
# G[2][0].remove('e1')
# del G[0][2]
# del G[2][0]
rec(G)
print(C)


## Cut-Edge Logic : Remove the edge and pass the recursion to all childs , requires connected-ness repetition calculation
## Tree Constructive Backtracking : Requires to check possibility of cycle after adding the edge to tree  
```
